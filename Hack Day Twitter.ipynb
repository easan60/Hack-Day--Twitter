```python
df_sent=df["text"]
df_sent=pd.DataFrame(df_sent)
```

```python
df_sent["text"]=df_sent["text"].str.lower()
df_sent["text"]=df_sent["text"].str.replace("[^\w\s]","")
df_sent["text"]=df_sent["text"].str.replace("\d+","")
df_sent["text"]=df_sent["text"].str.replace("\n"," ").replace("\r","")
```

```python
#Aksanlarƒ± ortadan kaldƒ±rdƒ±k
from unicodedata import normalize
df_sent["text"]=df_sent["text"].apply(lambda text:normalize("NFKD",str(text).encode("ascii","ignore").decode("utf-8","ignore")))
```

```python
#Gereksiz kelimeleri kaldƒ±rdƒ±k
from nltk.corpus import stopwords
stop_words=stopwords.words("english")
stop_words.append("rt") #rt kelimesi √ßok ge√ßtiƒüinden √ßƒ±kartƒ±ldu
df_sent['text']=df_sent['text'].apply(lambda x:" ".join([i for i in str(x).split(" ")  if i not in stop_words]))
```

```python
from langdetect import detect
df_sent["language"]=df_sent["text"].apply(detect)
df_sent=df_sent[df_sent["language"]=='en']
```

```python
df_sent.head()
```

```python
from textblob import TextBlob
df_sent[["polarity","subjectivity"]]=df_sent["text"].apply(lambda t:pd.Series(TextBlob(t).sentiment))
```

```python
df_sent.sample(4)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
df_sent['tokenized']=[tokenizer.tokenize(i) for i in df_sent['text'].values]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
df_sent['lemmatize']=[[lemmatizer.lemmatize(str(i)) for i in words] for words in df_sent['text'].values]
```

```python
df_sent.sample(3)
```

```python
df_sent.loc[df_sent['polarity']>0.5,'sentiment']=2 #pozitive
df_sent.loc[df_sent['polarity']==0.5,'sentiment']=1 #neutral
df_sent.loc[df_sent['polarity']<0.5,'sentiment']=0 #negative
df_sent.head()
```

```python
from wordcloud import WordCloud 
```

```python
def wc(data,bgcolor):
    from PIL import Image
    mask = np.array(Image.open("cloud.png"))
    wc=WordCloud(background_color=bgcolor, width=800, height=400,mask=mask,max_words=100).generate(' '.join(data))
    plt.figure( figsize=(20,10) )
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```

```python
possitive_data=df_sent[df_sent["subjectivity"]>0.5]["text"]
negative_data=df_sent[df_sent["subjectivity"]<0.5]["text"]
neutral_data=df_sent[df_sent["subjectivity"]==0.5]["text"]
```
